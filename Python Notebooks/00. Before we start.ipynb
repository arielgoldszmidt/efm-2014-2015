##What is Python##
Python is a general-purpose programming language that was built with the express purpose of being easy to read and simple to get things done with. It has a wide variety of uses including: web frameworks, data analysis, scientific programming, and even game development.

The easy-to-read part is important: if at any point you feel like your code is messy and you're jumping back and forth to make sense of it, FIX IT. Nipping the problem in the bud and keeping your code clean will pay off in the long run, especially when you try to export code you write for one project to another.

Python has several key characteristics:

1. <u>Interpreted, not compiled</u>: Unlike in languages like C or Java, Python code is interpreted, line-by-line, on the fly by the Python interpreter. In C, you compile your code into binaries (which is basically machine code), and when you try to run your program, the computer just opens up the binary and runs it. In Python, all code needs to be read through by a Python interpreter. Consequently, all problems with your code are only found when they are run through the interpreter. 
2. <u>Whitespace matters</u>: Again, unlike in languages like C or Java, where the syntax is largely indifferent to random spaces and linebreaks, in Python, spaces matter. (Note that this can be an issue when you use text editors that alternately use tabs or spaces.)
3. <u>Python does everything</u>: Python is an extremely flexible language because not only is it easy to use, there are libraries for EVERYTHING. Everything from web-scraping to web-frameworks to machine-learning to GUI libraries, you're bound to find a Python library that more or less does what you need to do.
4. <u>Python is slow</u>: The trade-off to having an easy to use language is the Python isn't terribly fast. For most uses, Python's speed should pose no problem, but when it comes to intensive data-analysis, Python's slow speed can be an issue. That said, more often than not, the time you save from the ease of writing code in Python alone will outweigh any computational speed loss.
5. <u>Python can be fast</u>: Because Python is slow, people have spent a lot of time writing libraries for Python that are blazingly fast. Libraries like NumPy and Scipy are written in C and Fortran to make up for Python's slow speed, and they are the bread-and-butter of any serious data analysis. In fact, because they're written in C, using NumPy and Scipy for data analysis easily outperform even R, a language specializing in statistical analysis.


So how do you use Python? There are generally two ways:

1. <u>Type commands directly into the Interpreter</u>: You can do this by opening up your terminal (or Command-prompt) and typing <u><b>python</b></u>. This should open the basic Python interpreter. Then you can go ahead and enter some command like:
2. <u>Give the Interpreter a script to run</u>: Instead of feeding the interpreter lines of code one by one, you can instead save your code in text form in a ".py" file, and run <u><b>python somescript.py</b></u>. That will run your code as if you fed it in line by line. You will generally use the former for testing some quick code, and the latter for running your data analysis. **This will change depending on your working style**


##What is IPython##
So you know that if you type <code>python</code> in terminal you start the python program that takes in lines and feeds it into the interpreter. Now think of that program (front-end) and the interpreter (back-end) as two entirely separate things: the text-based interface you see in your console is simply an interface feeding lines to some mysterious "server" working in the background.

**IPython** is basically an improved version of the text-based interface, that adds a whole bunch of really, really useful features. Among them are:
* Tab-based prediction of commands
* Pre-importing common libraries like NumPy and Matplotlib
* Integration with Matplotlib
* Ability to run python scripts on the fly
* Better error messages
* Pretty colors
And so on.

The important thing to take away is that IPython does not fundamentally change anything about the Python interpreter running in the background - it is simply an extremely good front-end change.

You can run IPython by typing <code>ipython</code> in your console. Have fun!

##What is IPython Notebook##
So we've finally come to this. IPython Notebook is a really, really cool feature that IPython has. Basically, starts a kind of server on your own computer, and creates a web-based environment for you to play around with Python and make Notebooks such as these. It's great for showcasing code in a presentable way.

IPython Notebook makes use of the underlying Python interpreter and allows you to run code in the middle of the page, such as this:

```python
print "Hello, World!"
```

What's even better is that you can edit that code on the fly, and re-run it.

```python
## Edit the following:
print "Edit me"
## Now press Shift+Enter
```

Now, isn't that fancy?

The purpose of using IPython Notebooks is that we can easily write Python notes, and you can play around with the code without having to go through the hassle of downloading and editting a script. These notebooks are saved in a ".ipynb" format, that you can easily export.
